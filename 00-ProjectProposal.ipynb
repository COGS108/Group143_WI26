## Research Question

**BROAD RQ:**

Which factors predict the outcome of dogs in the U.S. entering animal shelters, specifically whether a dog is adopted, transferred to another organization, or euthanized?

**SPECIFIC RQs:**

1. Do younger dogs have higher adoption probabilities than older dogs after accounting for breed and intake characteristics?
2. Are certain breeds of dogs significantly associated with adoption, transfer, or euthanasia outcomes?
3. Is an increased length of stay associated with a lower probability of adoption?



## Background and Prior Work

The purpose of animal shelters in the United States (U.S.) is to serve as a temporary place for animals to have a safe place to live while they wait for permanent homes. Humans can visit shelters, adopt an animal, and bring it home to care for. However, over the past decade, shelters have been overburdened and overcrowded, leading to many animals being euthanised [1, 2]. More specifically, 11.5% of dogs were euthanized in 2025 [2]. Dogs, in particular, require more human interaction, emphasizing the urgent need to study these species. The longer a dog stays in a shelter, the more likely they are to exhibit stress-related behaviors, such as aggression, which can reduce the likelihood of adoption [3]. Furthermore, dogs are active animals, and being confined to small spaces for long periods of time can be detrimental to their health [4].


One way to reduce the number of dogs euthanised is to increase the adoption rate. To do so, we aim to identify the best predictors to increase the likelihood of adoption. In a literature review, researchers identified the most common predictive factors affecting an animal’s adoption rate, including animal type, breed, color, gender, age, location, and length of stay [3]. Previous literature identified age, size, and color as the strongest predictive factors in a dog’s length of stay [5]. Other researchers found that age is a strong predictor of adoption, with younger dogs at intake in the shelter more likely to be adopted. Older dogs were more frequently taken in at the animal shelters [2, 6]. 


Our study focuses on shelters in the U.S., particularly in Austin, Texas; Sonoma, California; Louisville, Kentucky; and Bloomington, Indiana. According to PetFinder data, Texas has the highest rate of dog exports because shelters there cannot support the number of dogs without homes [7]. California has a medium-to-high level of exported dogs. Kentucky has a medium size of exported dogs, and Indiana has among the lowest rates of exported dogs. We wanted to explore the range of exported dogs based on animal shelter capacity to ensure we captured a diverse set of locations and shelter capacities.

## Hypothesis


1. We hypothesize that younger dogs and dogs without medical conditions will have a higher probability of adoption than older dogs and dogs with medical conditions. Alternatively, there could be no significant difference between age and medical condition and adoption rates, which would support the null hypothesis. 
2. Additionally, we expect longer stays in shelters to be associated with lower adoption probabilities, as dogs that stay in shelters for long periods may face increased behavioral and health challenges. 
3. We expect these patterns to hold even when comparing dogs of similar breeds and intake conditions.
